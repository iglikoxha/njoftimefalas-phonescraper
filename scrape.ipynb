```python
from bs4 import BeautifulSoup
from selenium import webdriver
from xlwt import Workbook
```

```python
# Open headless browser only once
options = webdriver.FirefoxOptions()
options.add_argument("--headless")
browser = webdriver.Firefox(options=options)

wb = Workbook()
sheet1 = wb.add_sheet('Nr. Kontaktit')

sheet_nr = 1

links = []

# open file in read mode
with open('njoftimefalas-links.csv', 'r', newline='') as read_obj:
    for row in read_obj:
       links.append(row.strip())

def scrape(link):
    global sheet_nr

    url = str(link)
    browser.get(url)

    html = browser.page_source

    soup = BeautifulSoup(html, "html.parser")

    num = soup.find(attrs={"class" : "track"})
    if num:
        num = num.attrs["href"][4:]
        
        date = soup.find(attrs={"class": "date"})
        if date: 
            date_val = date.find("small").find("b").text

        print(date_val)
        sheet1.write(sheet_nr, 0, num)
        sheet1.write(sheet_nr, 1, date_val)
        
        print(num)
        print(f"Jane bere {sheet_nr} iteracione.")

        sheet_nr = sheet_nr + 1
```

```python
from multiprocessing import Pool

if __name__ == "__main__":
    # 3 workers increase this for speeeeeed
    pool = Pool(3)
    pool.map(scrape, links[0:5])
    pool.close()

wb.save('njoftimefalas numbers.xls')
```
