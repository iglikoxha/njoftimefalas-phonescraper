```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
# Open headless browser only once
options = webdriver.FirefoxOptions()
options.add_argument("--headless")
driver = webdriver.Firefox(options=options)

for id in range(431630, 431640):
    url = f"https://www.njoftimefalas.com/23/oferta-vende-pune/id-{id}"
    driver.get(url) 

    html = driver.page_source

    soup = BeautifulSoup(html, "html.parser")

    cat = soup.find(attrs={"class" : "category"})
    if cat:
        if "Oferta Vende Pune" in cat.find("a").text:
            num = soup.find(id=id)
            if num:
                print(num.attrs["href"][4:])
```
